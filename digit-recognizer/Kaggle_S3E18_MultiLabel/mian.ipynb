```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("./train.csv")
test = pd.read_csv("./test.csv")

pd.set_option('display.max_columns', None)
print(f"There are {train.isnull().sum().sum()} null values in data set")
print("THe shape of train dataset is ", train.shape)
train.head()
```

```python
print("THe shape of test dataset is ", test.shape)
test.head()
```

```python
# print(train["EC1"].value_counts())
# print(train["EC2"].value_counts())

train.drop(["id", "EC3", "EC4", "EC5", "EC6"], inplace=True, axis=1)
# train.drop(["FpDensityMorgan1", "FpDensityMorgan2", "FpDensityMorgan3"], inplace=True, axis=1)
print(train[["EC1", "EC2"]].value_counts())
data_train = train
```

```python
# Data Balance

from sklearn.utils import resample

ec_01_minority = pd.DataFrame(train[(train["EC1"]==0) & (train["EC2"]==1)])
ec_10_minority = pd.DataFrame(train[(train["EC1"]==1) & (train["EC2"]==0)])
ec_00_minority = pd.DataFrame(train[(train["EC1"]==0) & (train["EC2"]==0)])
ec_11_majority = pd.DataFrame(train[(train["EC1"]==1) & (train["EC2"]==1)])

ec_01_minority = resample(ec_01_minority, replace=True,n_samples=7507, random_state=42)
ec_10_minority = resample(ec_10_minority, replace=True,n_samples=7507, random_state=42)
ec_00_minority = resample(ec_00_minority, replace=True,n_samples=7507, random_state=42)

data_train = pd.concat([ec_00_minority, ec_01_minority, ec_10_minority, ec_11_majority])
data_train.shape
```

```python
y = data_train[["EC1", "EC2"]]
data_train.drop(["EC1", "EC2"], inplace=True, axis=1)
print(y.head())
# y[["EC1", "EC2"]].value_counts()
y.shape
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(data_train, y, test_size=0.2, random_state=419)
```

```python
import optuna
from catboost import CatBoostClassifier
from optuna.samplers import TPESampler
from sklearn.multiclass import OneVsRestClassifier


def objective(trial):
    model = OneVsRestClassifier(CatBoostClassifier(
        iterations=trial.suggest_int("iterations", 100, 1000),
        learning_rate=trial.suggest_float("learning_rate", 1e-3, 1e-1, log=True),
        depth=trial.suggest_int("depth", 4, 10),
        l2_leaf_reg=trial.suggest_float("l2_leaf_reg", 1e-8, 100.0, log=True),
        bootstrap_type=trial.suggest_categorical("bootstrap_type", ["Bayesian"]),
        random_strength=trial.suggest_float("random_strength", 1e-8, 10.0, log=True),
        bagging_temperature=trial.suggest_float("bagging_temperature", 0.0, 10.0),
        od_type=trial.suggest_categorical("od_type", ["IncToDec", "Iter"]),
        od_wait=trial.suggest_int("od_wait", 10, 50),
        verbose=False
    ))
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    return accuracy_score(y_test, y_pred)


optuna.logging.set_verbosity(optuna.logging.WARNING)
sampler = TPESampler(seed=1)
study = optuna.create_study(study_name="catboost", direction="maximize", sampler=sampler)
print(study.optimize(objective, n_trials=100))

print("Number of finished trials: ", len(study.trials))
print("Best trial:")
trial = study.best_trial
print("  Value: ", trial.value)
print("  Params: ")
for key, value in trial.params.items():
    print("    {}: {}".format(key, value))

```
